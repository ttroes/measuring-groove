```python
i = 0
for i in range(cycles):

    # initialise variables 
    
    step1 = []
    step2 = []
    step3 = []
    step4 = []
    step5 = []
    step6 = []
    step7 = []
    step8 = []
    step9 = []
    step10 = []
    step11 = []
    step12 = []
    step13 = []
    step14 = []
    step15 = []
    step16 = []

    # AP-clustering exemplars
    step1 = np.random.choice(time, p=prob_1)
    step2 = np.random.choice(time, p=prob_2)
    step5 = np.random.choice(time, p=prob_5)
    step9 = np.random.choice(time, p=prob_9)
    step12 = np.random.choice(time, p=prob_12)
    step13 = np.random.choice(time, p=prob_13)
    
    print 'Exemplars selected'
    
    ### step 2: 2, 3, 7, 11, 15 ###
    step3 = np.random.choice(time, p=prob_3)
    while not (step2-ms_threshold) <= step3 <=(step2+ms_threshold):
        step3 =np.random.choice(time, p=prob_3)

    step7 = np.random.choice(time, p=prob_7)
    while not (step2-ms_threshold) <= step7 <=(step3+ms_threshold):
        step7 =np.random.choice(time, p=prob_7)

    step11 = np.random.choice(time, p=prob_11)
    while not (step2-ms_threshold) <= step11 <=(step3+ms_threshold):
        step11 =np.random.choice(time, p=prob_11)
    
    step15 = np.random.choice(time, p=prob_15)
    while not (step2-ms_threshold) <= step15 <=(step3+ms_threshold):
        step15 =np.random.choice(time, p=prob_15)
        
    print 'Step 2 done'
                
    ### step 12: 4,6,8,10,12,14,16 ###   
   
    step4 = np.random.choice(time, p=prob_4)
    #while not (step12-ms_threshold-20) <= step4 <=(step12+ms_threshold+20):
    #    step4 =np.random.choice(time, p=prob_4)
    print 'step 4'

    step6 = np.random.choice(time, p=prob_6)
    #while not (step12-ms_threshold-20) <= step6 <=(step12+ms_threshold+20):
    #    step6 =np.random.choice(time, p=prob_6)
    print 'step 6'
        
    step8 = np.random.choice(time, p=prob_8)
    while not (step12-ms_threshold) <= step8 <=(step12+ms_threshold):
        step8 =np.random.choice(time, p=prob_8)
    print 'step 8'

    step10 = np.random.choice(time, p=prob_10)
    while not (step12-ms_threshold) <= step10 <=(step12+ms_threshold):
        step10 =np.random.choice(time, p=prob_10)
    print 'step 10'
   
    step14 = np.random.choice(time, p=prob_14)
    while not (step12-ms_threshold) <= step14 <=(step12+ms_threshold):
        step14 =np.random.choice(time, p=prob_14)
        
    print 'step 14'
   
    step16 = np.random.choice(time, p=prob_16)
    while not (step12-ms_threshold) <= step16 <=(step12+ms_threshold):
        step16 =np.random.choice(time, p=prob_16)
        
    print 'Step 12 done'
        
    ids1.append(step1)
    ids2.append(step2)
    ids3.append(step3)
    ids4.append(step4)
    ids5.append(step5)
    ids6.append(step6)
    ids7.append(step7)
    ids8.append(step8)
    ids9.append(step9)
    ids10.append(step10)
    ids11.append(step11)
    ids12.append(step12)
    ids13.append(step13)
    ids14.append(step14)
    ids15.append(step15)
    ids16.append(step16)
    
    print 'One cycle is over'
    print step1, step2, step3,step4,step5,step6,step7,step8,step9,step10,step11,step12,step13,step14,step15,step16
    print '--'
    
    i += 1
```
