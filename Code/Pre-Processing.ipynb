```python
# function to produce a metrical grid, based on the first and last beat - needs beat count
def plotting_metrical_grid(filename, first_downbeat, last_four, counted_beats):
    load_file = 'sounds/drummer3/' + filename
    loader = es.MonoLoader(filename=load_file)
    audio = loader()

    last_beat = counted_beats * 4 - 4
    counted_beats = counted_beats * 4 - 3
    idealSixteenths = np.linspace(first_downbeat, last_four, counted_beats)
    diff_six = idealSixteenths[2]-idealSixteenths[1]
    for i in [1,2,3]:
        added_six = idealSixteenths[last_beat] + diff_six*i
        idealSixteenths = np.append(idealSixteenths, added_six)

    fs=44100
    t = np.arange(len(audio))/float(fs)
    plt.figure(figsize=(10,8))
    plt.plot(t, audio)
    for loc in idealSixteenths:
        plt.axvline(x=loc, color='y')
        
    #for i in range(len(beats)):
    #    plt.axvline(idealSixteenths[4*i], color = 'r')
```

```python
def microtiming_drummer3_thesis(filename, first_downbeat, last_four, counted_beats):
    read_onset = []
    onsets = []
    last_beat = []
    idealSixteenths = []
    
    load_file = 'sounds/drummer2/' + filename
    loader = es.MonoLoader(filename=load_file)
    audio = loader()

    last_beat = counted_beats * 4 - 4
    counted_beats = counted_beats * 4 - 3
    idealSixteenths = np.linspace(first_downbeat, last_four, counted_beats)
    diff_six = idealSixteenths[2]-idealSixteenths[1]
    for i in [1,2,3]:
        added_six = idealSixteenths[last_beat] + diff_six*i
        idealSixteenths = np.append(idealSixteenths, added_six)
    np.savetxt('grid_072_ideals.txt', idealSixteenths)
```
