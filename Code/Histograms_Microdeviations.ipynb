```python
%matplotlib inline
import numpy as np
import scipy
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def heatmap_draw(filename, title, plotname):
    var1 = pd.read_csv(filename)
    a = [0.100, 0.090, 0.080, 0.070, 0.060, 0.050, 0.040, 0.030, 0.020, 0.010, 0.005, 0, -0.005, -0.010, -0.020, -0.030, -0.040, -0.050, -0.060, -0.070, -0.080, -0.090, -0.100]
    plt.figure(figsize=(10,6))
    var1.index = a
    #cmap = sns.cubehelix_palette(rot=0, as_cmap = True)
    sns.heatmap(var1, cmap=sns.cubehelix_palette(8))
    plt.title(title)
    plt.xlabel('Ideal Sixteenths')
    plt.ylabel('Micro-deviation')
    plt.axvline(x=4, linestyle = ':')
    plt.axvline(x=8, linestyle = ':')
    plt.axvline(x=12, linestyle = ':')
    plt.axhline(y=12, linestyle = '-',c = 'red', alpha = 0.4)
    plt.axhline(y=11, linestyle = '-', c = 'red', alpha = 0.4)
    plt.axvline(x=0, ymax = 0.475, ymin = 0.525, linestyle = ':', c = 'red')
    plt.axvline(x=16, ymax = 0.475, ymin = 0.525, linestyle = ':', c = 'red')
    plt.savefig(plotname)
    plt.close()
```

```python
heatmap_draw('heatmaps/drummer1/heatmap_drummer1_allinstruments.csv', 'Drummmer1 - Phrases // Microdeviations around ideal sixteenths', 'drummer1_phrases.all_instruments.jpg')
```
