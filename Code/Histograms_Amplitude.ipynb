```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import IPython.display
import essentia
import essentia.standard as es
import librosa
import librosa.display
from utils import *
import pandas as pd
import seaborn as sns
```

```python
def heatmap_draw(filename, title, plotname):
    var1 = pd.read_csv(filename)
    a = [50,30,20,10,5,3,2.75,2.5,2.25,2,1.75,1.5,1.25,1,0.75,0.5,0.4,0.3,0.2,0.1,0.01,0.00001,0]
    plt.figure(figsize=(10,6))
    var1.index = a
    cmap = sns.cubehelix_palette(rot=-.1, as_cmap = True)
    sns.heatmap(var1, cmap = cmap)
    plt.title(title)
    plt.xlabel('Ideal Sixteenths')
    plt.ylabel('Amplitude')
    plt.axvline(x=4, linestyle = ':')
    plt.axvline(x=8, linestyle = ':')
    plt.axvline(x=12, linestyle = ':')
    plt.axhline(y=14, linestyle = '-',color = 'red', alpha = 0.4)
    plt.axhline(y=10, linestyle = '-', color = 'red', alpha = 0.4)
    plt.axhline(y=18, linestyle = '-', color = 'red', alpha = 0.4)
    plt.savefig(plotname)
    plt.close()
```

```python
heatmap_draw('heatmaps/drummer1/heatmap_amplitude_drummer1_allinstruments.csv', 'Drummmer1 - Phrases // Amplitude of ideal sixteenths', 'drummer1_phrases_amplitude.all_instruments.jpg')
```
